```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_function(x, ys, name: str):
    plt.plot(x, ys)
    plt.title(name)
    # plt.close()
    plt.show()
    
x = np.linspace(10, -10, 100)
x
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

plot_function(x, sigmoid(x), "Sigmoid")
```

```python
def step(x):
    return np.piecewise(x, [x < 0.0, x > 0.0], [0, 1])

plot_function(x, step(x), "step")
```

```python
def relu(x):
    return np.maximum(0, x)

plot_function(x, relu(x), "ReLu")
```

```python
def tanh(x):
    return np.tanh(x)

plot_function(x, tanh(x), "Tanh")
```

```python
def softmax(x):
    v_exp = np.exp(x)
    v_exp_sum = np.sum(v_exp)
    return [a / v_exp_sum for a in v_exp]

plot_function(x, softmax(x), "SoftMax")
```
